**5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

**Ans:-** A probability distribution function (PDF) gives the probability of a random variable taking on a specific value at a single point, while a cumulative distribution function (CDF) gives the probability that a random variable will be less than or equal to a certain value, essentially accumulating probabilities up to that point; in simpler terms, the PDF shows the "density" of probability at a specific point, while the CDF shows the "total probability accumulated" up to a certain point on the distribution curve.

**What is a PDF?**

PDF stands for Probability Density Function. It is an important concept in statistics for understanding probabilities related to continuous random variables. It is a smooth curve that shows how likely different outcomes are within a range of values.

For example, consider the temperature in a city on a given day. The PDF could show the likelihood of temperatures falling within certain ranges, like between 70°F and 80°F.

PDF does not give the probability of specific values, but rather the probability of the variable falling within a small interval around a particular value. The area under the PDF curve for a range of values represents the probability of the variable falling within that range. To find the probability of a single value, it requires to calculate the integral of the PDF at that point, which means finding the area under the curve at that specific value.

**What is a CDF?**

CDF stands for cumulative distribution function. The CDF complements the Probability Density Function and provides a cumulative view of the probabilities linked to a random variable. Unlike the smooth curve of the PDF, the CDF appears as a step function, jumping at specific values. It shows the probability that a random variable will be less than or equal to a given value.

Starting from 0 for negative values, the CDF gradually increases to 1 as the value of the random variable increases. For discrete random variables, the CDF rises in steps, corresponding to the probabilities of each possible outcome. With continuous random variables, it increases smoothly and reflects the combined probabilities across different intervals.

It is characterized by a single parameter, λ (lambda), which represents the average rate of occurrence of the event. The distribution is used when the events are rare and the number of occurrences is non-negative and can take on integer values (0, 1, 2, 3,…).

The key assumptions of the Poisson distribution are:

1. Events occur independently of each other.
2. The average rate of occurrence (λ) is constant over the given interval.
3. The number of events can be any non-negative integer.

Poisson Distribution Formula

      P (X = k) = e−λλk​ / k!

λ is the Average Rate of Occurrence of Events

k is the Number of Events that Occur

P(X = k) = (e-λ × λk) / k!

3) Moment Generating Function (MGF): MGF provides a way to derive moments of the distribution. It is represented by:
M(t) = e(λ(e^t – 1))

4) Characteristic Function (CF): CF is an alternative way to describe the distribution and is given by:
ϕ(t) = e(λ(e^(it) – 1))

5) Probability Generating Function (PGF): PGF generates the probabilities of the distribution and is expressed as:
G(z) = e(λ(z – 1))

2] Mode: Mode, or the most probable value, is simply the integer part of λ, denoted as ⌊λ⌋.

3] Mean and Variance: The mean (λ) and variance (λ) of a Poisson distribution are equal. This means that both the average number of events and the spread or variability around this average are characterized by the same parameter.

6) Non-negative and Discrete: The Poisson distribution describes the probability of non-negative integer values only, as it models counts of events. It is a discrete probability distribution.

7) Memorylessness: Events in a Poisson process are memoryless, meaning the probability of an event occurring in the future is independent of the past, given the current state. **13) What is the Central Limit Theorem (CLT), and why is it critical in statistics? **

**Ans:-** The Central Limit Theorem in Statistics states that as the sample size increases and its variance is finite, then the distribution of the sample mean approaches normal distribution irrespective of the shape of the population distribution.

The central limit theorem posits that the distribution of sample means will invariably conform to a normal distribution provided the sample size is sufficiently large. This holds regardless of the underlying distribution of the population, be it normal, Poisson, binomial, or any alternative distribution.


**Central Limit Theorem**


The Central Limit Theorem explains that the sample distribution of the sample mean resembles the normal distribution irrespective of the fact that whether the variables themselves are distributed normally or not. The Central Limit Theorem is often called CLT in abbreviated form.

When large samples usually greater than thirty are taken into consideration then the distribution of sample arithmetic mean approaches the normal distribution irrespective of the fact that random variables were originally distributed normally or not.


Central Limit Theorem Examples
Let’s say we have a large sample of observations and each sample is randomly produced and independent of other observations. Calculate the average of the observations, thus having a collection of averages of observations. Now as per the Central Limit Theorem, if the sample size is adequately large, then the probability distribution of these sample averages will approximate to a normal distribution.

Assumptions of the Central Limit Theorem
The Central Limit Theorem is valid for the following conditions:

1. The drawing of the sample from the population should be random.
2. The drawing of the sample should be independent of each other.
3. The sample size should not exceed ten percent of the total population when sampling is done without replacement.
4. Sample Size should be adequately large.
5. CLT only holds for a population with finite variance.

```python
#6) Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(x, low, high):

    if x < low:
        return 0
    elif x > high:
        return 1
    else:
        return (x - low + 1) / (high - low + 1)

# Define the range of the distribution
low = 1
high = 6  # Example: a standard die

# Generate x-values for the plot
x_values = np.arange(low - 1, high + 2)  # Include values outside the range

# Calculate CDF values for each x
cdf_values = [discrete_uniform_cdf(x, low, high) for x in x_values]

# Plot the CDF
plt.step(x_values, cdf_values, where='post')  # Use step plot for discrete CDF
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Values')
plt.ylabel('Cumulative Probability')
plt.xticks(x_values)  # Show all x-values on x-axis
plt.grid(True)  # Add grid for better visualization
plt.show()
```

```python
#19) Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
import scipy.stats as stats

# Sample data
sample_data = [48, 50, 52, 55, 58, 60, 62, 65, 68, 70]
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Use ddof=1 for sample standard deviation

# Population parameters (assumed)
population_mean = 55  # Null hypothesis value
population_std = 10  # Assumed population standard deviation



# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample_data)))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Set the significance level (alpha)
alpha = 0.05

# Make a decision
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.3f}")
```

```python
#20)  Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

# Sample data
data = [25, 30, 35, 38, 40, 42, 45, 47, 50, 52]

# Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation

# Set confidence level
confidence_level = 0.95  # 95% confidence interval

# Calculate the margin of error
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, len(data) - 1) * sample_std / np.sqrt(len(data))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Level: {confidence_level:.0%}")
print(f"Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# Interpretation
print("\nInterpretation:")
print(f"We are {confidence_level:.0%} confident that the true population mean lies within the interval "
      f"({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f}).")
```

```python
#22)  Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate x-values for the plot
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

# Calculate the probability density function (PDF) values
pdf = stats.norm.pdf(x, mu, sigma)

# Plot the PDF
plt.plot(x, pdf, color='blue', linewidth=2)
plt.title('Normal Distribution PDF (mu=0, sigma=1)')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
#23)  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for the Poisson distribution
mu = 3  # Average rate of events (lambda)

# Generate x-values (number of events)
x = np.arange(0, 15)  # Values from 0 to 14

# Calculate the cumulative distribution function (CDF)
cdf = poisson.cdf(x, mu)

# Print CDF values
for i in range(len(x)):
    print(f"P(X <= {x[i]}): {cdf[i]:.4f}")

# Plot the CDF
plt.step(x, cdf, where='post')  # Use step plot for discrete CDF
plt.title('Poisson Distribution CDF (mu=3)')
plt.xlabel('Number of Events (X)')
plt.ylabel('Cumulative Probability (P(X <= x))')
plt.grid(True)
plt.show()
```

```python
#24)  Simulate a random variable using a continuous uniform distribution and calculate its expected value.
import numpy as np

# Parameters for the continuous uniform distribution
low = 2  # Lower bound
high = 5  # Upper bound
num_samples = 10000  # Number of samples to generate

# Generate random samples from the uniform distribution
samples = np.random.uniform(low, high, num_samples)

# Calculate the expected value (theoretical)
expected_value_theoretical = (low + high) / 2

# Calculate the expected value (simulated)
expected_value_simulated = np.mean(samples)

# Print the results
print(f"Theoretical Expected Value: {expected_value_theoretical:.2f}")
print(f"Simulated Expected Value: {expected_value_simulated:.2f}")
```

```python
#25) Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets
np.random.seed(42)  # Set seed for reproducibility
dataset1 = np.random.normal(loc=50, scale=10, size=100)
dataset2 = np.random.normal(loc=60, scale=15, size=100)

# Calculate standard deviations
std_dev1 = np.std(dataset1, ddof=1)  # ddof=1 for sample standard deviation
std_dev2 = np.std(dataset2, ddof=1)

# Print standard deviations
print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

# Visualize the difference
plt.figure(figsize=(8, 6))
plt.hist(dataset1, bins=20, alpha=0.5, label='Dataset 1')
plt.hist(dataset2, bins=20, alpha=0.5, label='Dataset 2')
plt.title('Comparison of Standard Deviations')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

# Add vertical lines for standard deviations
plt.axvline(x=np.mean(dataset1) + std_dev1, color='blue', linestyle='--', label='Dataset 1 Std Dev')
plt.axvline(x=np.mean(dataset1) - std_dev1, color='blue', linestyle='--')
plt.axvline(x=np.mean(dataset2) + std_dev2, color='orange', linestyle='--', label='Dataset 2 Std Dev')
plt.axvline(x=np.mean(dataset2) - std_dev2, color='orange', linestyle='--')

plt.show()
```

```python
#26)  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np

# Generate data from a normal distribution
np.random.seed(42)  # Set seed for reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std Dev=10, 100 samples

# Calculate range
data_range = np.ptp(data)  # Using np.ptp() to get the range (peak-to-peak)

# Calculate interquartile range (IQR)
q1 = np.quantile(data, 0.25)  # 1st quartile (25th percentile)
q3 = np.quantile(data, 0.75)  # 3rd quartile (75th percentile)
iqr = q3 - q1

# Print the results
print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#27)  Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data (you can replace this with your own dataset)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate Z-scores
def zscore_normalize(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

normalized_data = zscore_normalize(data)

# Visualize the transformation
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Original data
sns.histplot(data, ax=axs[0], kde=True)
axs[0].set_title('Original Data')
axs[0].set_xlabel('Values')
axs[0].set_ylabel('Frequency')

# Normalized data
sns.histplot(normalized_data, ax=axs[1], kde=True)
axs[1].set_title('Z-score Normalized Data')
axs[1].set_xlabel('Z-scores')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#28) Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):

    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # Note: This calculates excess kurtosis (kurtosis - 3)

    return data_skewness, data_kurtosis

# Example usage:
np.random.seed(42)  # Set random seed for reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # Generate data from normal distribution

skewness, kurtosis = calculate_skewness_kurtosis(data)

print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```
